```python
import pandas as pd
import requests
import json

PAPAGO_URL = 'https://naveropenapi.apigw.ntruss.com/nmt/v1/translation'
PAPAGO_ID = ''
PAPAGO_SECRET = ''
```

```python
url='sootawon.csv'
df=pd.read_csv(url,sep=',',header=0)
df.head(5)
```

```python
df['trans_text']=""
```

```python
text=df.cont.values
```

```python
headers = {
    'X-NCP-APIGW-API-KEY-ID': PAPAGO_ID,
    'X-NCP-APIGW-API-KEY': PAPAGO_SECRET,
    'Content-Type': 'application/json'
}

data = {
    'source': 'ko',
    'target': 'en',
    'text': ""
    }
```

```python
i=0
for elem in text:
    elem=elem.replace("\r","")
    elem=elem.replace("\n",". ")
    elem=elem.replace("?"," ")
    data['text']=elem
    if(df['trans_text'][i]==""):
        response = requests.post('https://naveropenapi.apigw.ntruss.com/nmt/v1/translation', headers=headers, json=data)
        parsed_data=response.json()
        translated_text = parsed_data['message']['result']['translatedText']
        df['trans_text'][i]=translated_text 
    i+=1    
    
```

```python
trans=df['trans_text'].values

i=1
for elem in trans:
    print(i,":",elem)
    i+=1
```

```python
df.to_csv('sootawon_trans.csv')
```
